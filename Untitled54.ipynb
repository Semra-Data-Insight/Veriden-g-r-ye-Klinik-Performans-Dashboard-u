```python
import pandas as pd
import numpy as np
import pyodbc
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
from dateutil.relativedelta import relativedelta

# ==========================================
# 0) SQL bağlantısı (mevcut 'conn' varsa onu kullan)
# ==========================================
try:
    conn
except NameError:
    conn = pyodbc.connect(
        "DRIVER={ODBC Driver 17 for SQL Server};"
        "SERVER=SEMRA\\SQLEXPRESS;"
        "DATABASE=Klinik;"
        "Trusted_Connection=yes;"
    )

# ==========================================
# 1) Aylık TOPLAM CIRO ve BRÜT KÂR (Klinik / mart.klinik_ozet)
# ==========================================
sql = """
SELECT
    yil,
    ay,
    toplam_gelir   AS ciro,       
    toplam_kar      AS kar
FROM  mart.klinik_ozet;
"""
df = pd.read_sql(sql, conn)

# Basit tarih kolonu ve sıralama
df["tarih"] = pd.to_datetime(df["yil"].astype(str) + "-" + df["ay"].astype(str) + "-01")
df = df.sort_values("tarih").reset_index(drop=True)
# Aynı tarihleri birleştir, ortalama veya toplam al
df = df.groupby(["yil", "ay"], as_index=False)[["ciro", "kar"]].sum()
df["tarih"] = pd.to_datetime(df["yil"].astype(str) + "-" + df["ay"].astype(str) + "-01")
# (İsteğe bağlı) Eksik ayları 0 ile tamamlama — çizgiler kopmasın
full_idx = pd.date_range(df["tarih"].min(), df["tarih"].max(), freq="MS")
df = (
    df.set_index("tarih")
      .reindex(full_idx)
      .rename_axis("tarih")
      .reset_index()
)
df["yil"] = df["yil"].fillna(df["tarih"].dt.year)
df["ay"] = df["ay"].fillna(df["tarih"].dt.month)
df["ciro"] = df["ciro"].fillna(0.0)
df["kar"]  = df["kar"].fillna(0.0)

# Zaman endeksi (1..N)
df["zaman_indeksi"] = np.arange(1, len(df) + 1)

# ==========================================
# 2) MODELLER
# ==========================================
# 2a) CİRO (trend: zaman_indeksi)
X_ciro = df[["zaman_indeksi"]].values
y_ciro = df["ciro"].values
ciro_model = LinearRegression()           # istersen Ridge(alpha=1.0) yapabilirsin
ciro_model.fit(X_ciro, y_ciro)

ay_dummies = pd.get_dummies(df["ay"], prefix="ay", drop_first=True)
# 2b) KÂR (özellikler: zaman_indeksi + ciro)
X_kar_df = pd.concat([df[["zaman_indeksi", "ciro"]], ay_dummies], axis=1)
y_kar = df["kar"].values
kar_model = Ridge(alpha=1.0)              # biraz daha stabil; LinearRegression() da olur
kar_model.fit(X_kar, y_kar)

# Eğitim (in-sample) değerlendirme
y_kar_tahmin_ic = kar_model.predict(X_kar)
r2  = r2_score(y_kar, y_kar_tahmin_ic)
rmse = mean_squared_error(y_kar, y_kar_tahmin_ic, squared=False)
print(f"Kâr Modeli — R²: {r2:.3f} | RMSE: {rmse:,.0f}")

# ==========================================
# 3) GELECEK 3 AY TAHMİN (trend + ciro-bilgilendirmeli kâr)
# ==========================================
adim_sayisi = 3
son_idx = df["zaman_indeksi"].max()
gelecek_indeks = np.arange(son_idx + 1, son_idx + adim_sayisi + 1)  # t+1, t+2, t+3

# 3a) Gelecek 3 ay CİRO tahmini (saf trend)
gelecek_ciro = ciro_model.predict(gelecek_indeks.reshape(-1,1))

# (Opsiyonel) Yumuşatma: trend %70 + son 3 ay ortalaması %30
if len(df) >= 3:
    son3_ort = df["ciro"].tail(3).mean()
    gelecek_ciro = 0.7 * gelecek_ciro + 0.3 * son3_ort

# 3b) Gelecek 3 ay KÂR tahmini (özellikler: zaman_indeksi + forecast_ciro)
X_gelecek_kar = np.column_stack([gelecek_indeks, gelecek_ciro])
gelecek_kar_tahmin = kar_model.predict(X_gelecek_kar)

# Takvim etiketleri
son_tarih = df["tarih"].max()
gelecek_tarihler = pd.date_range(son_tarih + relativedelta(months=1), periods=adim_sayisi, freq="MS")

tahmin_df = pd.DataFrame({
    "tarih":  gelecek_tarihler,
    "zaman_indeksi": gelecek_indeks,
    "tahmin_ciro":   gelecek_ciro,
    "tahmin_kar":    gelecek_kar_tahmin
})
print("\n➡️ Gelecek 3 Ay Tahmini:")
print(tahmin_df[["tarih", "tahmin_ciro", "tahmin_kar"]])

# ==========================================
# 4) GRAFİK
# ==========================================
plt.figure(figsize=(8,4.8))
plt.plot(df["zaman_indeksi"], df["kar"], label="Gerçek Kâr", marker="o")
plt.plot(df["zaman_indeksi"], y_kar_tahmin_ic, label="Model (İç Uyum)", linestyle="--", alpha=0.7)
plt.plot(tahmin_df["zaman_indeksi"], tahmin_df["tahmin_kar"], label="Tahmin (3 Ay)", marker="x", linestyle="--")
plt.title("Aylık Brüt Kâr — Trend + Ciro Bilgili Model")
plt.xlabel("Zaman (Ay)")
plt.ylabel("Brüt Kâr")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ==========================================
# 5) Power BI için CSV (TR uyumlu)
# ==========================================
csv_yol="klinik_kar_son3ay_tr.csv"
tahmin_df[["tarih", "tahmin_kar"]].to_csv(
    "klinik_kar_son3ay_tr.csv",
    index=False,
    float_format="%.2f",
    encoding="utf-8-sig",
    sep=",",      # sadece virgül ayraç (Power BI otomatik tanır)
    decimal=","   # ondalık nokta
)
print(f"\nCSV kaydedildi: {csv_yol}")
```
